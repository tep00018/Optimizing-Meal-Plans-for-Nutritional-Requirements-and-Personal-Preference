    * Population: 100
    * Generations: 10
    * Parent Selection: Tournament, k=5
    * Crossover: Uniform Crossover, p_cross=0.25
    * Mutation: p_mut = 0.2
    * Survivor Selection: Best of parent, child, mutant

```python
#################################################################
# Nutrition Fitness Function 
#################################################################
def fitness(v):
    rating = []
    cal = []
    fat = []
    pro = []
    carb = []
    fiber = []
    sugar = []
    chol = []
    satfat = []
    netcarb = []
    vita = []
    niacin = []
    vitb6 = []
    vitc = []
    folate = []
    calcium = []
    iron = []
    magn = []
    pot = []
    sod = []
    thiamin = []
    # Define nutrient parameters for food list
    for i in range(len(v)):
        rating.append(user1List.loc[user1List['cms_id'] == v[i]]['rating'])
        cal.append(user1List.loc[user1List['cms_id'] == v[i]]['calories'])
        fat.append(user1List.loc[user1List['cms_id'] == v[i]]['fatContent'])
        pro.append(user1List.loc[user1List['cms_id'] == v[i]]['proteinContent'])
        carb.append(user1List.loc[user1List['cms_id'] == v[i]]['carbohydrateContent'])
        fiber.append(user1List.loc[user1List['cms_id'] == v[i]]['fiberContent'])
        sugar.append(user1List.loc[user1List['cms_id'] == v[i]]['sugarContent'])
        chol.append(user1List.loc[user1List['cms_id'] == v[i]]['cholesterolContent'])
        satfat.append(user1List.loc[user1List['cms_id'] == v[i]]['saturatedFatContent'])
        netcarb.append(netcarb)
        vita.append(user1List.loc[user1List['cms_id'] == v[i]]['vitamin a iu'])
        niacin.append(user1List.loc[user1List['cms_id'] == v[i]]['niacin'])
        vitb6.append(user1List.loc[user1List['cms_id'] == v[i]]['vitamin b6'])
        vitc.append(user1List.loc[user1List['cms_id'] == v[i]]['vitamin c'])
        folate.append(user1List.loc[user1List['cms_id'] == v[i]]['folate'])
        calcium.append(user1List.loc[user1List['cms_id'] == v[i]]['calcium'])        
        iron.append(user1List.loc[user1List['cms_id'] == v[i]]['iron'])
        magn.append(user1List.loc[user1List['cms_id'] == v[i]]['magnesium'])
        pot.append(user1List.loc[user1List['cms_id'] == v[i]]['potassium'])
        sod.append(user1List.loc[user1List['cms_id'] == v[i]]['sodium'])
        thiamin.append(user1List.loc[user1List['cms_id'] == v[i]]['thiamin'])       

    #Calculate total and percent values for food list
    total_cal = round(np.sum(cal), 1)
    total_fat = round(np.sum(fat), 1)
    total_pro = round(np.sum(pro), 1)
    total_carb = round(np.sum(carb), 1)
    total_fiber = round(np.sum(fiber), 1)
    total_sugar = round(np.sum(sugar), 1)
    total_chol = round(np.sum(chol), 1)
    total_satfat = round(np.sum(satfat), 1)
    total_netcarb = round((total_carb - total_fiber), 1)
    mean_rating = round(np.mean(rating), 2)
    total_vita = round(np.sum(vita), 1)
    total_niacin = round(np.sum(niacin), 2)
    total_vitb6 = round(np.sum(vitb6), 2)
    total_vitc = round(np.sum(vitc), 2)
    total_folate = round(np.sum(folate), 2)
    total_calcium = round(np.sum(calcium), 2)
    total_iron = round(np.sum(iron), 2)
    total_magn = round(np.sum(magn), 2)
    total_pot = round(np.sum(pot), 2)
    total_sod = round(np.sum(sod), 2)
    total_thiamin = round(np.sum(thiamin), 2)      
    p_cal = round(total_cal / target_cal, 2)
    p_vita = round(total_vita / 2333, 2)
    p_niacin = round(total_niacin / 14, 2)
    p_vitb6 = round(total_vitb6 / 1.3, 2)
    p_vitc = round(total_vitc / 75, 2)
    p_folate = round(total_folate / 400, 2)
    p_calcium = round(total_calcium / 1000, 2)
    p_iron = round(total_iron / 18, 2)
    p_magn = round(total_magn / 320, 2)
    p_pot = round(total_pot / 2600, 2)
    p_sod = round(total_sod / 1500, 2)
    p_thiamin = round(total_thiamin / 1.1, 2)
    n_fiber = round(total_fiber / 25, 2)

    #Calculate NNR with each value truncated at 100%
    nnr = round(((min(1, p_vita) + min(1, p_niacin) + min(1, p_vitb6) + min(1, p_vitc) + 
                  min(1, p_folate) + min(1, n_fiber) + min(1, p_calcium) + min(1, p_iron) + 
                  min(1, p_magn) + min(1, p_pot) + min(1, p_sod) + min(1, p_thiamin))/12), 2)

    #Calculate Upper Bound Penalty###############################################################
    # Sodium penalty
    if total_sod > 2300:
        sod_pen = round(np.absolute(((np.absolute(total_sod - 2300)**2./3.))/3526465), 5)
    else:
        sod_pen = 0
    # Cholesterol penalty
    if total_chol > 300:
        chol_pen = round(np.absolute(((np.absolute(total_chol - 300)**2./3.))/59952), 5)
    else:
        chol_pen = 0
    # Saturated Fat penalty
    if total_satfat > 0.1 * target_cal/9:
        satfat_pen = round(np.absolute(((np.absolute(total_satfat - 0.1 * target_cal/9)**2./3.))/533), 5)
    else:
        satfat_pen = 0
    # Sugar penalty
    if total_sugar > 40:
        sugar_pen = round(np.absolute(((np.absolute(total_sugar - 40)**2./3.))/1066), 5)
    else:
        sugar_pen = 0       
        
    # Sum of penalty functions
    ubp = sod_pen + chol_pen + satfat_pen + sugar_pen
    
    #Calorie Penalty Function (allows up to 110% of target then applies a penalty)################
    if total_cal < 0.8*target_cal or total_cal > 1.2*target_cal:
        cal_pen = 0
    else:
        cal_pen = 1  

    #Rating Penalty Function  ################
    rating_pen = -0.05 * (mean_rating * mean_rating) + (0.55 * mean_rating) - 0.5

    score = round((nnr - ubp) * (cal_pen) * (rating_pen), 4)
    return score
```

```python
#################################################################
# Initial Population (Random Generation)
#################################################################
def unsorted_population(pop_size, target_cal):
    for j in range(pop_size):
        breakfast1 = random.choice(breakfast)
        lunch1 = random.choice(lunch)
        dinner1 = random.choice(dinner)
        side1 = random.choice(side)
        side2 = random.choice(side)
        snack1 = random.choice(snack)
        snack2 = random.choice(snack)
        dessert1 = random.choice(dessert)
        v = [breakfast1, lunch1, dinner1, side1, side2, snack1, snack2, dessert1]
        food = []
        cal = []
        for i in range(len(v)):
            cal.append(user1List.loc[user1List['cms_id'] == v[i]]['calories'])
            total_cal = np.sum(cal)
            if total_cal < target_cal:
                food.append(v[i])
            else:
                food.append(999999)
        x = fitness(v=food)
        total_cal==0
        x_list.append(x)
        food_list.append(food)
    unsorted_pop = dict(zip(x_list, food_list))
    return unsorted_pop
```

```python
#################################################################
# Tournament Selection, Random (k = 5)
#################################################################
def tournament_selection(population, pop_size, target_cal):
    mum_key = np.max(random.choices(list(population.keys()), weights=None, cum_weights=None, k=5))
    mum = population[mum_key]
    dad_key = np.max(random.choices(list(population.keys()), weights=None, cum_weights=None, k=5))
    dad = population[dad_key]
    return mum, dad, mum_key, dad_key
```

```python
#################################################################
# Crossover (Uniform, cross_prob = probability of cross @ each gene) - 2 children
#################################################################
def uniform_crossover(population, pop_size, target_cal, cross_prob):
    mum = tournament_selection(population, pop_size, target_cal)[0]
    dad = tournament_selection(population, pop_size, target_cal)[1]
    mum_key = tournament_selection(population, pop_size, target_cal)[2]
    dad_key = tournament_selection(population, pop_size, target_cal)[3]
    child1 = copy.deepcopy(mum)
    child2 = copy.deepcopy(dad)
    for i in range(len(mum)):
        if random.random() < cross_prob:
            child1[i], child2[i] = dad[i], mum[i]
    child1_key = fitness(child1)
    child2_key = fitness(child2)
    return mum, dad, child1, child2, mum_key, dad_key, child1_key, child2_key
```

```python
#################################################################
# Mutation (Random according to mutation probability)
#################################################################
def fitness_mutation1(population, pop_size, target_cal, cross_prob, mut_prob):
    h = uniform_crossover(population, pop_size, target_cal, cross_prob)
    mum = h[0]
    child1 = h[2]
    mum_key = h[4]
    child1_key = h[6]
    mutant1 = copy.deepcopy(child1)
    for i in range(len(mutant1)):
        if i == 0:
            if random.random() < mut_prob:
                mutant1[0] = random.choice(breakfast)
        elif i == 1:
            if random.random() < mut_prob:
                mutant1[1] = random.choice(lunch)
        elif i == 2:
            if random.random() < mut_prob:
                mutant1[2] = random.choice(dinner)
        elif i == 3:
            if random.random() < mut_prob:
                mutant1[3] = random.choice(side)
        elif i == 4:
            if random.random() < mut_prob:
                mutant1[4] = random.choice(side)
        elif i == 5:
            if random.random() < mut_prob:
                mutant1[5] = random.choice(snack)
        elif i == 6:
            if random.random() < mut_prob:
                mutant1[6] = random.choice(snack)
        elif i == 7:
            if random.random() < mut_prob:
                mutant1[7] = random.choice(dessert)
    # Compare and keep best of mum, child1 and mutant1
    mutant1_key = fitness(mutant1)
    if mum_key > child1_key and mum_key > mutant1_key:
        offspring1 = mum
        offspring1_key = mum_key
    elif child1_key > mum_key and child1_key > mutant1_key:
        offspring1 = child1
        offspring1_key = child1_key
    else:
        offspring1 = mutant1
        offspring1_key = mutant1_key
    return offspring1, offspring1_key
        
def fitness_mutation2(population, pop_size, target_cal, cross_prob, mut_prob):
    h = uniform_crossover(population, pop_size, target_cal, cross_prob)
    dad = h[1]
    child2 = h[3]
    dad_key = h[5]
    child2_key = h[7]
    mutant2 = copy.deepcopy(child2)
    for i in range(len(mutant2)):
        if i == 0:
            if random.random() < mut_prob:
                mutant2[0] = random.choice(breakfast)
        elif i == 1:
            if random.random() < mut_prob:
                mutant2[1] = random.choice(lunch)
        elif i == 2:
            if random.random() < mut_prob:
                mutant2[2] = random.choice(dinner)
        elif i == 3:
            if random.random() < mut_prob:
                mutant2[3] = random.choice(side)
        elif i == 4:
            if random.random() < mut_prob:
                mutant2[4] = random.choice(side)
        elif i == 5:
            if random.random() < mut_prob:
                mutant2[5] = random.choice(snack)
        elif i == 6:
            if random.random() < mut_prob:
                mutant2[6] = random.choice(snack)
        elif i == 7:
            if random.random() < mut_prob:
                mutant2[7] = random.choice(dessert)
    # Compare and keep best of dad, child2 and mutant2
    mutant2_key = fitness(mutant2)
    if dad_key > child2_key and dad_key > mutant2_key:
        offspring2 = dad
        offspring2_key = dad_key
    elif child2_key > dad_key and child2_key > mutant2_key:
        offspring2 = child2
        offspring2_key = child2_key
    else:
        offspring2 = mutant2
        offspring2_key = mutant2_key
    return offspring2, offspring2_key
```

```python
#################################################################
# GA Flow for Population Size and Generations (Fitness Driven)
# Run several generations and output best solution
#################################################################
def genetic_alg(iterations, generations, pop_size, cross_prob, mut_prob):
    pairs = int(pop_size / 2) # Number of breeding pairs in each generation
    genm_dict = {} # Dictionary to log best individual for each generation  (key = fitness value, value = best individul)
    lead_dict = {} # Dictionary to log best fitness value for each generation  (key = generation, value = best fitness)
    plot_dict = {} # Dictionary to log all fitness value for each generation (key = generation, value = all fitness values)
    trial_dict = {} # Dictionary of data from each iterations (key = iteration, value = leader for each generation)
    best3_dict = {} # Dictionary of best fitness, vector from each generation (key = fitness value, value = associated individul)
    print()
    print("{:<5} {:<4} {:<8} {:<91} {:<7}".format("Iter", "Gen", "Fitness", "Leader Meal Vector", "Ex Time"))
    print("{:<5} {:<4} {:<8} {:<91} {:<7}".format("----", "---", "-------", "-----------------------------------------------------------------------------------------", "-------"))   
    for m in range(1, iterations+1):
        #print(f'Iteration: {m}')
        # Trial Header
        for i in range(1, generations+1):
            startTime = time.time() #Start iteration timer
            if i == 1: 
                population = unsorted_population(pop_size, target_cal) #Gen 1 starting population (Gen 0) is randomly generated
            else: 
                population = genm_pop #Gen 2+ starting population are offspring of previous generation
            # Reset variables
            off1fitness = []
            off2fitness = []
            offspring_list = []
            fitness_list = []
            pop_dict = {} # resets pop to gather next generation of offspring
            lead = [] # leader vector in offspring population
            lead_key = [] # fitness of lead vector
            lead_key_list = []
            lead_list = []
            for j in range(pairs): # Loops for each breeding pair
                x = fitness_mutation1(population, pop_size, target_cal, cross_prob, mut_prob)
                offspring1 = x[0] # First offspring generated
                off1fitness = x[1] # Fitness of first offspring calculated
                offspring_list.append(offspring1) # Creates list of all offspring in generation
                fitness_list.append(off1fitness) # Creates list of all offspring fitness values in generation
                y = fitness_mutation2(population, pop_size, target_cal, cross_prob, mut_prob)
                offspring2 = y[0]
                off2fitness = y[1] # Fitness of second offspring calculated
                offspring_list.append(offspring2) # Creates list of all offspring in generation
                fitness_list.append(off2fitness)  # Creates list of all offspring fitness values in generation
            pop_dict = dict(zip(fitness_list, offspring_list)) # Generational dictionary for value: offspring and key: fitness 
            lead_key = np.max(list(pop_dict.keys())) # Selects top individual fitness from population
            lead = pop_dict[lead_key] # Top individual from population
            lead_dict[i] = lead_key # Dictionary to log best fitness value for each generation  (key = gernation, value = best fitness)
            lead_key_list.append(lead_key)
            plot_dict[i] = fitness_list # Dictionary to log all fitness value for each generation (key = gernation, value = all fitness values)
            genm_dict[lead_key] = lead # Dictionary to log best individual for each generation  (key = generation, value = best individul)
            exTime = round((time.time() - startTime),2)
            print("{:<5} {:<4} {:<8} {:<91} {:<7}".format(m, i, round(lead_key,4), str(lead), exTime))
            genm_pop = pop_dict #stores current population to use as parents for next generationand (pop_dict gets reset)

            # Save Best 3
            sorted_dict = sorted(pop_dict, reverse=True)
            for k in (sorted_dict[0:3]):
                best3_dict[k] = pop_dict[k]
        # Output 
        # Plot Leader for each Generation
        #plt.plot(lead_dict.values())
        #plt.title('Genetic Algorithm Leader Fitness by Generation')
        #plt.show()

        # Plot Generation Distribution
        #plt.boxplot(plot_dict.values())
        #plt.title('Genetic Algorithm Population Fitness by Generation')
        #plt.show()

        # Output Best Menu from last generation ############################################################################
        leader_key = np.max(list(genm_pop.keys())) # Identifies best fitness
        leader = genm_pop[leader_key]  # Identifies best individual
        #decode(leader) # Menu output for best individual
        trial_dict[m] = lead_dict
    return leader, leader_key
```
